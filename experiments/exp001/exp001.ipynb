## Power analysis

In the previous section, we mentioned the possibility of doing either a full or fractional $2^7$ factorial design.
I'll explore these options using simulation and see what the tradeoffs are between sample size and power.
I found a nice package called [pyDOE2](https://pypi.org/project/pyDOE2/) that is able to generate factorial experiment designs for you automatically, so we'll use that to help us figure out the power of our potential experments.

```python
import itertools as it
import warnings
from contextlib import ExitStack

import pyDOE2 as doe
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
from statsmodels.stats.anova import anova_lm
import matplotlib.pyplot as plt
import tqdm
```

We can also explore fractional factorial designs.
`pyDOE2` has several functions for producing fractional designs according to different criteria.
I'll use `fracfact_by_res`, which generates designs with different "resolutions" (the minimum number of characters in the "generator" string, see [these notes](https://www.itl.nist.gov/div898/handbook/pri/section3/pri3344.htm) for more details); the higher the resolution, the better we'll be able to unconfound higher-order interaction effects.
I don't think any effects of greater than 3rd-order are going to be important, so resolution III is probably the highest we'll need, but we can consider higher resolutions for the higher effective replication factor.

```python
for r in reversed(range(3, 7)):
    num_samples = len(doe.fracfact_by_res(7, r))
    print(f"Resolution {r}: {num_samples} samples")
```

We could also try a Plackett-Burman design, which has intermediate numbers of samples (multiples of 4), but results in partial confounding between some effects.

```python
design = doe.pbdesign(4) # main effects
f = lambda i, j: (design[:,i] * design[:, j])[:,None]
design = np.concatenate([design, f(0, 1), f(0, 2), f(0, 3), f(1, 2), f(1, 3), f(2, 3)], axis=1)
np.corrcoef(design.T)
```

Now we need to create a simulation of the treatment assignment, collection of experimental data (with a synthetic treatment effect), and analysis so that we can estimate the power.

```python
def gen_data(
    contrasts: np.ndarray,
    treatment_effects: dict[str, float],
    baseline: float,
    noise: float,
) -> pd.DataFrame:
    """Function for generating synthetic data given an experiment design, treatment effects, etc."""
    design = (contrasts > 0).astype(int)
    num_samples, num_factors = design.shape
    factor_labels = [chr(ord("A") + i) for i in range(num_factors)]
    df = pd.DataFrame(data=design, columns=factor_labels)
    sum_treatment_effects = np.zeros((num_samples,))
    for label, effect in treatment_effects.items():
        treated_units = df[list(label)].prod(axis=1) == +1
        sum_treatment_effects += effect * treated_units
    obs = baseline + sum_treatment_effects + noise * np.random.randn(num_samples)
    df["obs"] = obs
    # df[factor_labels] = contrasts
    return df

df = gen_data(
    contrasts=doe.fracfact_by_res(7, 6),
    # contrasts=doe.fold(doe.ff2n(4)),
    treatment_effects={"AB": +3},
    baseline=0.0,
    noise=1.0,
)
df.head()
```

```python
def analyze(data: pd.DataFrame, max_interaction_order: int=1):
    """Analyzes the experiment data and returns some results."""
    num_samples = len(data)
    formula = "obs ~ "
    # add main effects
    treatment_labels = list(data.columns.difference(["obs"]))
    if len(treatment_labels) >= num_samples:
        raise ValueError(f"More main effects ({len(treatment_labels)}) than there are samples ({num_samples})!")
    formula += " + ".join(treatment_labels)
    # add interaction effects
    interaction_terms = [":".join(L) for L in it.chain(*(it.combinations(treatment_labels, r) for r in range(2, max_interaction_order+1)))]
    if len(treatment_labels) + len(interaction_terms) >= num_samples:
        warnings.warn("More model parameters than there are samples; consider lowering max_interaction_order")
    formula += ("" if len(interaction_terms) == 0 else " + ") + " + ".join(interaction_terms)
    # fit model
    model = smf.ols(formula, data).fit()
    # fit a null model so we can get a likelihood ratio statistic (test for significance at the entire model level)
    null_model = smf.ols("obs ~ 1", data).fit()
    log_likelihood_ratio = null_model.llf - model.llf
    # perform ANOVA (tests for significance at individual factor level)
    anova_df = anova_lm(model)
    return {
        "model": model,
        "anova_results": anova_df,
        "log_likelihood_ratio": log_likelihood_ratio,
    }

model, adf, llr = analyze(df, 2).values()
print(f"log likelihood ratio = {llr}")
adf[adf["PR(>F)"] < 0.05]
```

Let's put it all together into one convenient function.
Then we can use it to estimate the distribution of test statistics under the null hypothesis (no treatment effects), and use that to calculate p-values from those statistics.
I'll fix the `noise` parameter to 1, so you can think of the units as multiples of the standard deviation of the observation variable.

```python
# collect test statistics under null hypothesis to get p-values
def run_simulation(design: np.ndarray, num_iter: int, treatment_effects: dict[str, float], stat_name: str, analyze_kwargs={}, progress=True):
    """Runs a simulation of the experiment to collect test statistics"""
    statistics = []
    with ExitStack() as stack:
        if progress:
            pbar = stack.enter_context(tqdm.tqdm(total=num_iter))
        for _ in range(num_iter):
            data = gen_data(design, treatment_effects, baseline=0.0, noise=1.0)
            results = analyze(data, **analyze_kwargs)
            statistics.append(results[stat_name])
            if progress:
                pbar.update()
    statistics = sorted(statistics)
    return statistics

def pval_func_factory(null_statistics: list[float]):
    null_statistics = np.array(sorted(null_statistics))
    return lambda stat: np.interp(stat, null_statistics, np.linspace(0, 1, len(null_statistics)))

design = doe.fracfact_by_res(n=7, res=6)
null_llr_1 = run_simulation(design, num_iter=1000, treatment_effects={}, stat_name="log_likelihood_ratio", analyze_kwargs={"max_interaction_order": 1})
get_pvalue_1 = pval_func_factory(null_llr_1)
null_llr_2 = run_simulation(design, num_iter=1000, treatment_effects={}, stat_name="log_likelihood_ratio", analyze_kwargs={"max_interaction_order": 2})
get_pvalue_2 = pval_func_factory(null_llr_2)
```

```python
fig, axes = plt.subplots(ncols=2, figsize=(2*7, 5))

# histogram of log likelihood ratio
plt.sca(axes[0])
plt.hist(null_llr_1, bins=30, density=True, histtype="step", label="main effects")
plt.hist(null_llr_2, bins=30, density=True, histtype="step", label="main effects + interactions")
plt.xlabel("log likelihood ratio")
plt.ylabel("density")
plt.legend()

# mapping from log likelihood ratio to p-value 
plt.sca(axes[1])
llr = np.linspace(null_llr_1[0], null_llr_1[-1], 100)
pvalue = [get_pvalue_1(x) for x in llr]
plt.plot(llr, pvalue, label="main effects")
llr = np.linspace(null_llr_2[0], null_llr_2[-1], 100)
pvalue = [get_pvalue_2(x) for x in llr]
plt.plot(llr, pvalue, label="main effects + interactions")
plt.xlabel("log likelihood ratio")
plt.ylabel("p-value")
plt.legend()

plt.suptitle("Distribution of log likelihood ratio test statistic under null hypothesis", y=0.95)
plt.show()
```

Now we can simulate our experiment under various _alternative_ hypotheses (nonzero treatment effects) for a variety of full/fractional factorial experiment designs.
(I initially tried out some Plackett-Burman designs, but I think I misunderstood how they worked.)

```python
# set up configuration and simulation parameters to iterate over
num_iter_null, num_iter_alt = 1000, 100 # number of iterations for calculating the statistics under the null/alternative hypotheses
num_effects = 10  # the number of treatment effect sizes to try
stat_name = "log_likelihood_ratio"
num_factors = 7
simulation_params = list(it.product(
    ("A", "AB"), # effect type/order
    it.chain( # experiment designs
        ((doe.ff2n(num_factors), "full factorial"),), # full factorial
        ((doe.fracfact_by_res(num_factors, resolution), "fractional factorial") for resolution in range(3, 6+1)), # fractional factorial
    )
))

# construct p-value functions by simulating null hypothesis
print("generating null samples to calculate p-values")
treatment_effects, get_pvalue = {}, {} # no effect!
with tqdm.tqdm(total=len(simulation_params)) as pbar:
    for effect_type, (design, design_type) in simulation_params:
        num_samples = len(design)
        analyze_kwargs = {"max_interaction_order": 2 if num_samples > num_factors + num_factors*(num_factors-1)/2 else 1}
        null_statistics = run_simulation(design, num_iter_null, treatment_effects, stat_name, analyze_kwargs, progress=False)
        get_pvalue[effect_type, design_type, num_samples] = pval_func_factory(null_statistics)
        pbar.update()
# simulate alternative hypotheses and estimate power
print("generating samples under various alternatives to calculate power")
power_df = []
with tqdm.tqdm(total=num_effects * len(simulation_params)) as pbar:
    for effect_type, (design, design_type) in simulation_params:
        num_samples = len(design)
        analyze_kwargs = {"max_interaction_order": 2 if num_samples > num_factors + num_factors*(num_factors-1)/2 else 1}
        mio = analyze_kwargs["max_interaction_order"]
        for effect_magnitude in np.geomspace(1e-1, 1e1, num_effects):
            treatment_effects = {effect_type: effect_magnitude}
            statistics = run_simulation(design, num_iter_alt, treatment_effects, stat_name, analyze_kwargs, progress=False)
            pfunc = get_pvalue[effect_type, design_type, num_samples]
            for significance in [0.01, 0.05, 0.10]:
                power = np.mean([pfunc(stat) <= significance for stat in statistics])
                power_df.append((effect_type, effect_magnitude, design_type, num_samples, mio, significance, power))
            pbar.update()
power_df = pd.DataFrame(power_df, columns=["effect_type", "effect_magnitude", "design_type", "num_samples", "model_order", "sig_level", "power"])
```

```python
df = power_df.query("sig_level == 0.05 & abs(effect_magnitude - 1) < 0.5 & effect_type == 'A'")
df.sort_values(["effect_magnitude", "num_samples"], ascending=[True, True]).head(20)
```

```python
df = power_df.query("design_type == 'fractional factorial' & num_samples == 64 & power >= 0.6 & sig_level == 0.05")
merge_df = (
    df
    .groupby("effect_type")["effect_magnitude"].min()
    .to_frame()
    .reset_index()
)
df.merge(merge_df, on=["effect_type", "effect_magnitude"], how="inner")
```

```python
def lin_reg_inference(exp_data: pd.DataFrame, bag_data: pd.DataFrame):
    # pre-process the data
    df = exp_data.merge(bag_data, on=["bag num", "brand"], how="inner")
    df["kernel_age_months"] = (df["timestamp"] - df["purchased at"]) / timedelta(days=30)
    df["expansion_factor"] = df["popped volume (cup)"] / df["kernel mass (g)"]
    df["neg_sqrt_inedible_mass_ratio"] = -np.sqrt(df["inedible mass (g)"] / df["popped mass (g)"])
    df = df.rename(columns={
        "temp. profile": "temp_profile",
        "temp. magnitude": "temp_magnitude",
        "kernel qty": "kernel_qty",
        "steam release": "steam_release",
        "bag num": "bag_num",
        "burn rating (*/10)": "burn_rating",
        "overall quality (*/10)": "overall_quality",
    })
    # train the model
    cat_cols = [
        "brand", "temp_profile", "temp_magnitude", "oil", "kernel_qty", "agitation",
        "steam_release", #"bag_num",
        # don't use bag_num until there I have more individual bags, because it is
        # degenerate with brand
    ]
    num_cols = ["kernel_age_months"]
    features = '+'.join(f'C({c})' for c in cat_cols) + "+" + '+'.join(num_cols)
    for target in (
        "expansion_factor", "neg_sqrt_inedible_mass_ratio", "overall_quality", "burn_rating",
    ):
        formula = f"{target} ~ {features}"
        result = ols(formula, df[cat_cols + num_cols + [target]]).fit()
        # display results
        print(f"### Results for {target} ###")
        with warnings.catch_warnings():
            warnings.filterwarnings("ignore", "kurtosistest")
            display(result.summary().tables[1])

lin_reg_inference(exp_data, bag_data)
```
